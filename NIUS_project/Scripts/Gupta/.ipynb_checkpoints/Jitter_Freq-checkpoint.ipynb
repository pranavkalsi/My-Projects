```python
import os
import psrchive
import numpy as np
import matplotlib.pyplot as plt
from uncertainties import ufloat
from uncertainties import unumpy
```

```python
freqs = []

for i in range(split):
    load = f"fits/freq_{split}/493_446.000{i}_0000.fits"
    data = psrchive.Archive_load(load)
    data.fscrunch()
    freqs.append(data.get_centre_frequency())
    dump = f"fits/freq_{split}/493_1_{i}.fits"
    data.unload(dump)
```

```python
arrtim = psrchive.ArrivalTime()
arrtim.set_shift_estimator("PGS")  # can change to PGS, FDM (MCMC), COF, GIS
arrtim.set_format("Tempo2")

std = psrchive.Archive_load("fits/template.fits")
arrtim.set_standard(std)
```

```python
toas = []

for i in range(split):
    obs = psrchive.Archive_load(f"fits/freq_{split}/493_1_{i}.fits")
    arrtim.set_observation(obs)

    # obs.tscrunch()
    # obs.unload(f"fits/freq_{split}/493_1_{i}.tt")

    # std = psrchive.Archive_load(f"fits/freq_{split}/493_1_{i}.tt")
    # arrtim.set_standard(std)

    toa = arrtim.get_toas()
    toas.append(toa)
```

```python
for i in range(split):
    f = open(f"tim/freq_{split}/{i}.tim", "w")
    f.write("FORMAT 1\n")
    for j in range(len(toas[i])):
        f.write(toas[i][j] + "\n")
f.close()
```

```python
residuals = []

for i in range(split):
    residual = []
    ! tempo2 -f J1136+1551.par tim/freq_{split}/{i}.tim -residuals
    f = open("residuals.dat", "r")
    for line in f:
        residual.append([float(line.split()[1]), float(line.split()[2])])
    residuals.append(residual)
```

```python
i = 0
xvalues = np.arange(len(residuals[i]))
yvalues = np.array(residuals[i])[:,0]
yerr = np.array(residuals[i])[:,1]

slope, intercept = np.polyfit(xvalues, yvalues, 1)
yline = intercept + slope*xvalues

plt.errorbar(xvalues, yvalues, yerr, fmt=".")
plt.plot(xvalues, yline)
plt.xlabel("Subints")
plt.ylabel("Residuals (s)")
```

```python
valids = []

for i in range(len(residuals)):
    temp = np.array(residuals[i])
    valid = [1]*len(temp)
    timings = temp[:,0]
    
    mad = np.median(np.absolute(timings - np.median(timings)))
    # slope, intercept = np.polyfit(np.arange(len(temp)), timings, 1)
    x = np.median(timings)

    # stdev = 1.4826*MAD, 3*MAD = 2*stdev = 95% of data
    for k in range(len(timings)):
        # y = intercept + slope*k
        if abs(x-timings[k]) > 3*mad:
            valid[k] = 0
    valids.append(valid)

print([valids[i].count(1) for i in range(len(valids))])  
```

```python
for i in range(len(residuals)):
    temp = np.array(residuals[i])
    timings = temp[:,1]

    mad = np.median(np.absolute(timings - np.median(timings)))
    x = np.median(timings)

    # stdev = 1.4826*MAD, 4.5*MAD = 3*stdev = 99.7% of data
    for k in range(len(timings)):
        if abs(x-timings[k]) > 4.5*mad:
            valids[i][k] = 0

print([valids[i].count(1) for i in range(len(valids))])
```

```python
i = 0
tim = residuals[i]
index = [j for j in range(len(tim)) if valids[i][j] == 1]

xvalues = np.arange(len(tim))[index]
yvalues = np.array(tim)[:,0][index]
yerr = np.array(tim)[:,1][index]

slope, intercept = np.polyfit(xvalues, yvalues, 1)
yline = intercept + slope*xvalues

plt.errorbar(xvalues, yvalues, yerr, fmt=".")
plt.plot(xvalues, yline)
plt.xlabel("Subints")
plt.ylabel("Residuals (s)")
```

```python
for i in range(split):
    f = open(f"tim/freq_{split}/{i}_clean.tim", "w")
    f.write("FORMAT 1\n")
    for j in range(len(toas[i])):
        if valids[i][j] == 1:
            f.write(toas[i][j] + "\n")
f.close()
```

```python
for i in range(split):
    ! tempo2 -f J1136+1551.par tim/freq_{split}/{i}_clean.tim -fit F0 -residuals
    path = f"tim/freq_{split}/{i}_residuals.dat"
    os.rename("residuals.dat", path)
```

```python
rms = []

for i in range(split):
    residual_value = []
    residual_error = []

    f = open(f"tim/freq_{split}/{i}_residuals.dat", "r")
    for line in f:
        line = line.split()
        residual_value.append(float(line[1]))
        residual_error.append(float(line[2]))

    residual_value = np.array(residual_value)
    residual_error = np.array(residual_error)
    
    for t in range(len(residual_value)):
        residual_value[t] = abs(residual_value[t])

    arr = residual_value.argsort()
    residual_value = residual_value[arr[::1]][:350]
    residual_error = residual_error[arr[::1]][:350]

    residuals = unumpy.uarray(residual_value, residual_error)
    
    for j in range(len(residuals)):
        residuals[j] = residuals[j]**2
    
    rms.append(residuals.sum()/len(residuals))

print(rms)
```

```python
radiometer = []
for i in range(split):
    residual_error = []

    f = open(f"tim/freq_{split}/{i}_residuals.dat", "r")
    for line in f:
        line = line.split()
        residual_error.append(float(line[2]))

    residual_error = np.array(residual_error)
    mean = residual_error.sum()/len(residual_error)

    sum_new = 0
    for i in range(len(residual_error)):
        sum_new += (residual_error[i] - mean)**2
    sum_new /= len(residual_error)
    sum_new = sum_new**0.5
    
    x = ufloat(mean, sum_new)
    radiometer.append(x**2)

print(radiometer)
```

```python
jitter = []
for i in range(split):
    x = (rms[i] - radiometer[i])**0.5
    y = (1/1200)**0.5
    jitter.append(x*y)

print(jitter)
```

```python
xvalues = freqs
yvalues = []
yerr = []
for i in range(split):
    yvalues.append((jitter[i].nominal_value)*1e6)
    yerr.append((jitter[i].std_dev)*1e6)

f = open(f"tim/freq_{split}/{split}.txt", "w")
f.write("Freq       Jitter (us)        Jitter Error (us)\n")
for i in range(split):
    value = f"{xvalues[i]} {yvalues[i]} {yerr[i]} \n"
    f.write(value)
f.close()

plt.errorbar(xvalues, yvalues, yerr=yerr, fmt=".")
plt.xlabel("Frequency (MHz)")
plt.ylabel("Jitter (us)")
```
